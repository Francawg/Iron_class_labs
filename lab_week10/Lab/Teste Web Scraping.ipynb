# Importando bibliotecas

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

## Pegando a Url para webscraping
### para isso usamos o requests

```python
response = requests.get('http://www.adorocinema.com/series-tv/melhores/')
```

```python
response.content
```

## Criando a Sopa

```python
soup = BeautifulSoup(response.content)
```

## extraindo um nome do site 

```python
soup
```

```python
soup.find(name = 'a' , attrs = {'class' : 'meta-title-link'}).text
```

```python
soup.find(name = 'a', attrs = {'class':'meta-title-link', 'href':"/series/serie-3517/"}).text
```

### Chamando todos os nomes principais da página, quando fazemos isso nos retorna uma lista e nao um dicionário

```python
soup.find_all(name = 'a', attrs = {'class':'meta-title-link'})
```

```python
soup.find_all(name = 'a', attrs = {'class':'meta-title-link'})[-4].text
```

```python
#criando um looping para pegar todos os nomes das series

series = []

for i in range(15):
    series_name = soup.find_all(name = 'a', attrs = {'class':'meta-title-link'})[i].text
    series.append(series_name)
    
print(series)
```
